```python
#Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the data in the dataframe
df = pd.read_csv('employee_data.csv', encoding='latin1')
```

```python
#Renaming the columns for better access
df.columns = [column.lower().replace(' ','_') for column in df.columns]
```

```python
df.head()
```

```python
#Changing datatype to datetime
df['hire_date'] = pd.to_datetime(df['hire_date'])

#Changing datatype to datetime
df['exit_date'] = pd.to_datetime(df['exit_date'])

#Removing '$' and ',' from the salary column and converting it to int 
df['annual_salary'] = df['annual_salary'].astype(str)
df['annual_salary'] = df['annual_salary'].str.replace(',', '')
df['annual_salary'] = df['annual_salary'].str.replace('$', '')
df['annual_salary'] = df['annual_salary'].astype(int)

#removing '%' from bonus column and converting it to int
df['bonus_%'] = df['bonus_%'].astype(str)
df['bonus_%'] = df['bonus_%'].str.replace('%','')
df['bonus_%'] = df['bonus_%'].astype(int)
```

Now, while calculating duplicated values, we must have some domain knowledge regarding the dataset. In this case, the dataset contains employee information of all employees. We must know that while some job titles might be repeated, they cannot be considered duplicates as many people can have the same job. We have to identify a column which has unique value for each employee, which in this case is the employee id (eeid). So while checking duplicate values we can use the subset argument as eeid in the duplicated() and drop_duplicates() methods.

```python
df.duplicated(subset=['eeid'], keep='first')
```

Let us count the number of duplicate values

```python
df.duplicated(subset=['eeid'], keep='first').sum()
```

Now we drop the duplicate values using drop_duplicates() method and using eeid as value for the subset argument

```python
df = df.drop_duplicates(subset=['eeid'], keep='first')
```

Now let us check the shape of the data set once again

```python
df.shape
```

This tells us that duplicate values have been dropped as the number of rows changed from 1000 to 911 (meaning 89  duplicate values have been dropped)

### Conclusion: Duplicate values in the employee id column have been removed
